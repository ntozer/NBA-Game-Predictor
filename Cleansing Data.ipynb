```python
# standard imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\nathan.tozer\\Downloads\\NBA-Prediction-Analysis-master\\NBA-Prediction-Analysis-master\\data\\raw\\NBA_2011\\games-november.csv')
df = df.drop(['Notes', 'Box Score'], axis=1)

df['Day'] = df['Date'].str.split(' ').str[0]

day_dict = {'Sun': 1, 'Mon': 2, 'Tue': 3, 'Wed': 4, 'Thu': 5, 'Fri': 6, 'Sat': 7}
df['Day_Num'] = df['Date'].str.split(' ').str[0]
df['Day_Num'].replace(day_dict, inplace=True)

df.head(10)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
time_dif = df['Date'][20] - df['Date'][1]
print(time_dif)
```

```python
df['Spread'] = df['Home_PTS'] - df['Visitor_PTS']
df.head(10)
```

```python
x = df['Day_Num'].tolist()
y = df['Spread'].tolist()

plt.scatter(x, y, c='b', marker='x', label='Spread')
plt.show()
```
